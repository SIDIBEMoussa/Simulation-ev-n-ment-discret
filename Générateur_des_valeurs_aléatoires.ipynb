```python
from matplotlib.ticker import *
import matplotlib.pyplot as plt
import numpy as np
from simple_colors import *

def noncongruentiel1(f,X0):
    generate_list=[X0]
    generate_set=set(generate_list)
    while len(generate_list)==len(generate_set):
        X1=f(X0)
        generate_list.append(X1)
        generate_set=set(generate_list)
        X0=X1
    #Element determinant la période et supprimer cette redondance

    ele=generate_list.pop()
    
    # Index de cet élement
    
    m=generate_list.index(ele)
    
    # La période 
    
    l=len(generate_list)-m
    generate_list.append(ele)
    
    print("="*150)
    print(green("Le nombre de vmettre en grasaleur généré avant d'atteindre les éléments de la période m={}".format(m),"bold"))
    print(green("La période de la fonction génératrice l={}".format(l),"bold"))
    print(green("L'element Xp=Yp est:{}".format(generate_list[-1]),"bold"))
    print("="*150)

    #Observation graphigue

    plt.rcParams.update({"font.size":18})
    plt.figure(figsize=(12,8))
    plt.plot(np.linspace(m,m+l,l+1),generate_list[m:],"r-->")
    plt.axis(xmin=m-1,xmax=m+l+1)
    plt.axes().xaxis.set_major_locator(MultipleLocator(1))
    plt.grid()
    p=plt.hlines(y=generate_list[-1],xmin=m,xmax=m+l,colors="blue",linestyles="-")
    plt.title("Répresentation des elements générateur de la période")

    return generate_list,m,l
```

```python
#Exemple1
generate_list,m,l=noncongruentiel1(f,33)
```
